### Data Cleaning

```python
# drop the last 6 columns
df.drop(['Gross tertiary education enrollment (%)','Population','Unemployment rate','Urban_population','Latitude','Longitude'] ,axis=1, inplace= True)
```

```python
# replace some missing values with not available 
df[['category','Country','Abbreviation','channel_type','video_views_rank','country_rank','channel_type_rank','video_views_for_the_last_30_days','subscribers_for_last_30_days','created_year','created_month','created_date']] = df[['category','Country','Abbreviation','channel_type','video_views_rank','country_rank','channel_type_rank','video_views_for_the_last_30_days','subscribers_for_last_30_days','created_year','created_month','created_date']].fillna('Not Available')
```

```python
df.isnull().sum()
```

```python
# Replacing missing value with the mode (for integer dataType)
df[[]] = df[[]].fillna[].mode()[0]
```

```python
# drop remaining missing values
df.dropna(inplace=True)
```

```python
# change data column to date type

# Assuming 'created_date' is the name of your date column
df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce')

# Now, any date values that couldn't be parsed will be set to NaT
```
